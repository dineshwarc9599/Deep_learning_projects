```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Copy data
df_fs = df.copy()

# Handle missing values
num_cols = df_fs.select_dtypes(include=['int64','float64']).columns
df_fs[num_cols] = df_fs[num_cols].fillna(df_fs[num_cols].median())

cat_cols = df_fs.select_dtypes(include='object').columns
df_fs[cat_cols] = df_fs[cat_cols].fillna('Unknown')

# Define X and y
y = df_fs['Net Sale Value']
X = df_fs.drop(['Net Sale Value','Posting Date'], axis=1)

# Encoding
X = pd.get_dummies(X, drop_first=True)

# Train Fast Model
model = RandomForestRegressor(
    n_estimators=30,
    random_state=42,
    n_jobs=-1
)

model.fit(X, y)

# Get importance
importance = pd.Series(model.feature_importances_, index=X.columns)

# Convert Back To Original Column

original_importance = {}

for col in importance.index:
    
    # Get original column name
    if "_" in col:
        base_col = col.split("_")[0]
    else:
        base_col = col
    
    if base_col not in original_importance:
        original_importance[base_col] = 0
        
    original_importance[base_col] += importance[col]

# Convert to Series
original_importance = pd.Series(original_importance)

# Sort
original_importance = original_importance.sort_values(ascending=False)

print("Top Original Columns:")
print(original_importance.head(10))
```

```python
# Final Columns Used

mandatory_cols = ['Size','Range','Neck','Collar','Sleeve']

top_cols = ['Mrp','Style','Season','Qty','Warehouse Name']

final_cols = mandatory_cols + top_cols + ['Net Sale Value','Posting Date']

df_final = df[final_cols].copy()

print("Final Columns Used:")
print(df_final.columns)
```

```python
import numpy as np

def clean_season(x):
    
    x = str(x).upper()
    
    if "DIW" in x:
        return "Festive"
    
    elif "TKS" in x or "OFFER" in x:
        return "Promotion"
    
    elif "FAS" in x:
        return "Regular"
    
    elif "WF" in x:
        return "Winter"
    
    elif "AMJ" in x or "JFM" in x:
        return "Quarter"
    
    else:
        return "Other"

df_final['Season_Clean'] = df_final['Season'].apply(clean_season)

print(df_final['Season_Clean'].value_counts())
```

# **Time Series Forecasting**

```python
df_final['Posting Date'] = pd.to_datetime(df_final['Posting Date'])

df_final['Week_Start_Date'] = df_final['Posting Date'] - pd.to_timedelta(
    df_final['Posting Date'].dt.weekday, unit='D'
)
```

```python
weekly_df = df_final.groupby(
    ['Week_Start_Date','Size','Sleeve','Neck','Collar','Range'],
    as_index=False
)['Net Sale Value'].sum()
```

```python
weekly_df
```

```python
weekly_df['Net Sale Value'].value_counts()
```

```python
weekly_df = weekly_df.sort_values('Week_Start_Date')
```

```python
group_cols = ['Size','Sleeve','Neck','Collar','Range']

weekly_df['smooth_sales'] = (
    weekly_df.groupby(group_cols)['Net Sale Value']
    .transform(lambda x: x.rolling(4, min_periods=1).mean())
)
```

```python
weekly_df['smooth_sales'].value_counts()
```

```python
weekly_df['ewm_4'] = (
    weekly_df.groupby(group_cols)['Net Sale Value']
    .transform(lambda x: x.ewm(span=4, adjust=False).mean())
)
```

```python
weekly_df['is_zero'] = (weekly_df['Net Sale Value'] == 0).astype(int)
```

```python
weekly_df['lag_1'] = (
    weekly_df.groupby(group_cols)['smooth_sales']
    .shift(1)
)
```

```python
weekly_df["lag_52"] = weekly_df["Net Sale Value"].shift(52)
weekly_df["rolling_12"] = weekly_df["Net Sale Value"].rolling(12).mean()
```

```python
# Rolling mean (4 week)
weekly_df['rolling4'] = (
    weekly_df.groupby(group_cols)['Net Sale Value']
    .transform(lambda x: x.rolling(window=4).mean())
)

#  8 week rolling
weekly_df['rolling8'] = (
    weekly_df.groupby(group_cols)['Net Sale Value']
    .transform(lambda x: x.rolling(window=8).mean())
)
```

```python
# Fill NaN only for numeric columns
num_cols = weekly_df.select_dtypes(include=['number']).columns
weekly_df[num_cols] = weekly_df[num_cols].fillna(0)
```

```python
events = [
    {"Event": "Women's Day",      "From": "2024-03-08", "To": "2024-03-10"},
    {"Event": "Anniversary Sale", "From": "2024-09-04", "To": "2024-09-16"},
    {"Event": "Diwali",           "From": "2024-10-08", "To": "2024-10-31"},
    {"Event": "Women's Day",      "From": "2025-03-07", "To": "2025-03-09"},
    {"Event": "Anniversary Sale", "From": "2025-08-26", "To": "2025-09-07"},
    {"Event": "Diwali",           "From": "2025-09-27", "To": "2025-10-20"}
]
```

```python
weekly_df['Event_Flag'] = 0

for event in events:
    start = pd.to_datetime(event['From'])
    end   = pd.to_datetime(event['To'])
    
    weekly_df.loc[
        (weekly_df['Week_Start_Date'] >= start) &
        (weekly_df['Week_Start_Date'] <= end),
        'Event_Flag'
    ] = 1
```

```python
# Time features
weekly_df['Month'] = weekly_df['Week_Start_Date'].dt.month
weekly_df['Quarter'] = weekly_df['Week_Start_Date'].dt.quarter
weekly_df['Week'] = weekly_df['Week_Start_Date'].dt.isocalendar().week
weekly_df['Year'] = weekly_df['Week_Start_Date'].dt.year

# Trend feature
weekly_df['Trend'] = weekly_df['Year'] * 100 + weekly_df['Week']
```

```python
weekly_df['Event_Flag_Lag1'] = weekly_df.groupby(group_cols)['Event_Flag'].shift(1)
weekly_df['Event_Flag_Lag1'].fillna(0, inplace=True)
```

# **LightGBM**

```python
weekly_df = weekly_df.sort_values(['Week_Start_Date'])

train_size = int(len(weekly_df) * 0.8)

train = weekly_df.iloc[:train_size]
test  = weekly_df.iloc[train_size:]

X_train = train.drop(['Net Sale Value','Week_Start_Date'], axis=1)
y_train = train['Net Sale Value']

X_test  = test.drop(['Net Sale Value','Week_Start_Date'], axis=1)
y_test  = test['Net Sale Value']


from sklearn.preprocessing import LabelEncoder

cat_cols = ['Size','Sleeve','Neck','Collar','Range']

for col in cat_cols:
    le = LabelEncoder()
    
    X_train[col] = X_train[col].astype(str)
    X_test[col]  = X_test[col].astype(str)
    
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col]  = le.transform(X_test[col])
```

```python
from lightgbm import LGBMRegressor

model = LGBMRegressor(
    n_estimators=1500,
    learning_rate=0.01,
    num_leaves=64,   
    min_child_samples=50,  
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred = np.maximum(y_pred, 0)


from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

wape = (np.sum(np.abs(y_test - y_pred)) / np.sum(y_test)) * 100

print("MAE:", mae)
print("RMSE:", rmse)
print("WAPE:", wape)
```

# **XGBoost Regressor**

```python
weekly_df = weekly_df.sort_values(['Week_Start_Date'])

train_size = int(len(weekly_df) * 0.8)

train = weekly_df.iloc[:train_size]
test  = weekly_df.iloc[train_size:]

X_train = train.drop(['Net Sale Value','Week_Start_Date'], axis=1)
y_train = train['Net Sale Value']

X_test  = test.drop(['Net Sale Value','Week_Start_Date'], axis=1)
y_test  = test['Net Sale Value']


from sklearn.preprocessing import LabelEncoder

cat_cols = ['Size','Sleeve','Neck','Collar','Range']

for col in cat_cols:
    le = LabelEncoder()
    
    X_train[col] = X_train[col].astype(str)
    X_test[col]  = X_test[col].astype(str)
    
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col]  = le.transform(X_test[col])

# MODEL BUILDING

from xgboost import XGBRegressor

model = XGBRegressor(
    n_estimators=800,
    learning_rate=0.02,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_train, y_train)

# PREDICTION
y_pred = model.predict(X_test)

# EVALUATION
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# WAPE
wape = (np.sum(np.abs(y_test - y_pred)) / np.sum(y_test)) * 100

print("MAE:", mae)
print("RMSE:", rmse)
print("WAPE:", wape)
```

# **Time Series Cross Validation and Hyperparameter Tuning**

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
import numpy as np
```

```python
tscv = TimeSeriesSplit(n_splits=5)
```

```python
xgb_model = XGBRegressor(
    objective='reg:squarederror',
    random_state=42,
    n_jobs=-1
)
```

```python
param_dist = {
    'n_estimators': [1000, 2000, 3000],
    'learning_rate': [0.01, 0.005],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 0.1, 0.3]
}

random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=20,
    scoring='neg_root_mean_squared_error',
    cv=tscv,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

best_xgb = random_search.best_estimator_

print("Best Parameters:")
print(random_search.best_params_)
```

```python
y_pred = best_xgb.predict(X_test)
y_pred = np.maximum(y_pred, 0)

from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
wape = (np.sum(np.abs(y_test - y_pred)) / np.sum(y_test)) * 100

print("MAE:", mae)
print("RMSE:", rmse)
print("WAPE:", wape)
```
